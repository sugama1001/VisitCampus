```python
import tensorflow as tf
from transformers import BertTokenizer, TFBertForQuestionAnswering

model_name = 'cahya/bert-base-indonesian-522M'
tokenizer = BertTokenizer.from_pretrained(model_name)
model = TFBertForQuestionAnswering.from_pretrained(model_name)
```

```python
import pandas as pd
import os

# Ganti 'nama_file.csv' dengan nama file CSV yang ingin Anda baca
file_path = '/content/chatbot dataset - Final Dataset.csv'

# Membaca file CSV menjadi DataFrame
df = pd.read_csv(file_path)

# Menampilkan beberapa baris pertama dari DataFrame
print(df.head())
```

```python
input_ids = tokenizer(list(df['Pertanyaan']),list(df['Text']), return_tensors='tf', padding=True, truncation=True)
print(input_ids['input_ids'].numpy(),list(df['Jawaban']))
print(input_ids)
```

```python
import numpy as np

def find_subarray_indexes(array, subarray):
    n = len(array)
    m = len(subarray)

    for i in range(n - m + 1):
        if np.array_equal(array[i:i+m], subarray):
            return i, i + m - 1

    return -1, -1  # Return -1, -1 jika subarray tidak ditemukan

dataset_y = []
# Contoh penggunaan
main_array = input_ids['input_ids'].numpy()
sub_array = list(df['Jawaban'])

for x in range(len(main_array)):
    konteks = main_array[x]
    answer = tokenizer(sub_array[x], return_tensors='tf',add_special_tokens=False)['input_ids'].numpy()[0]
    # print(konteks)
    # print(answer)
    start_index, end_index = find_subarray_indexes(konteks, answer)

    if start_index != -1:
        dataset_y.append([start_index, end_index])
        print(f"Subarray ditemukan dari indeks {start_index} hingga {end_index}.")
    else:
        dataset_y.append([0, 0])
        print("Subarray tidak ditemukan.")

print(dataset_y)
```

```python
indeks_00 = [index for index, item in enumerate(dataset_y) if item == [0, 0]]
dataset_y = [item for item in dataset_y if item != [0, 0]]
print(indeks_00)
print(dataset_y)
```

```python
def question_answering_loss(y_true, y_pred):
    print(y_true,y_pred)
    # Extract start_positions and end_positions from y_true
    start_positions = y_true[:, 0]  # Assuming the start_positions are in the first column
    end_positions = y_true[:, 1]    # Assuming the end_positions are in the second column

    # Extract start_logits and end_logits from y_pred
    start_logits = y_pred[:, :tf.shape(y_pred)[-1] // 2]
    end_logits = y_pred[:, tf.shape(y_pred)[-1] // 2:]
    print(start_logits,end_logits)

    start_positions_one_hot = tf.one_hot(start_positions, depth=tf.shape(start_logits)[-1])
    end_positions_one_hot = tf.one_hot(end_positions, depth=tf.shape(end_logits)[-1])

    start_loss = tf.keras.losses.categorical_crossentropy(start_positions_one_hot, start_logits, from_logits=True)
    end_loss = tf.keras.losses.categorical_crossentropy(end_positions_one_hot, end_logits, from_logits=True)

    # Sum both losses
    total_loss = start_loss + end_loss

    return total_loss
```

```python
def build_classifier_model():
    input_shape = (None,)
    input_ids = tf.keras.layers.Input(shape=input_shape, dtype=tf.int32, name='input_ids')
    token_type_ids = tf.keras.layers.Input(shape=input_shape, dtype=tf.int32, name='token_type_ids')
    attention_mask = tf.keras.layers.Input(shape=input_shape, dtype=tf.int32, name='attention_mask')
    output_neurons = input_shape[0]
    input_layer = {
        'attention_mask': attention_mask,
        'input_ids': input_ids,
        'token_type_ids': token_type_ids
    }

    # Mengambil output dari layer BERT
    output = loaded_model(input_layer)
    return tf.keras.Model(inputs=input_layer, outputs=output)
```

```python
model = build_classifier_model()
model.summary()
```

```python
def custom_accuracy_loss(y_true, y_pred):
    # Menghitung akurasi
    accuracy = tf.reduce_mean(tf.cast(tf.equal(tf.argmax(y_true, axis=-1), tf.argmax(y_pred, axis=-1)), tf.float32))

    # Menghitung loss berdasarkan akurasi (1 - akurasi)
    loss = 1 - accuracy

    return loss
```

```python
model.compile(optimizer=optimizer, loss=question_answering_loss)
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint

# Path untuk menyimpan checkpoint pada setiap epoch
checkpoint_path_epoch = "/content/drive/MyDrive/chatbot_epoch.h5"

# Path untuk menyimpan checkpoint model terbaik
checkpoint_path_best = "/content/drive/MyDrive/chatbot_best.h5"

# Callback untuk menyimpan checkpoint pada setiap epoch
checkpoint_callback_epoch = ModelCheckpoint(
    filepath=checkpoint_path_epoch,
    save_weights_only=True,
    save_best_only=False,  # Set ke False agar menyimpan pada setiap epoch
    verbose=1
)

# Callback untuk menyimpan checkpoint model terbaik
checkpoint_callback_best = ModelCheckpoint(
    filepath=checkpoint_path_best,
    save_weights_only=True,
    save_best_only=True,  # Set ke True agar hanya menyimpan model terbaik
    monitor='loss',
    mode='min',
    verbose=1
)
```
